```python
import pandas as pd
```

```python
tfidf = pd.read_csv("validation_results/val_tfidf.csv",index_col=0)
tfidf
```

```python
svr = pd.read_csv("validation_results/val_svr.csv",index_col=0)
svr
```

```python
ngram = pd.read_csv("validation_results/val_ngram.csv",index_col=0)
ngram
```

```python
roberta = pd.read_csv("validation_results/val_bert.csv",index_col=1)
roberta = roberta.iloc[: , 1:]
roberta
```

```python
#df_list = [tfidf, svr, ngram, roberta]
df1 = pd.DataFrame()
df1 = pd.concat([tfidf['cohesion'],svr['cohesion'],ngram['cohesion'],roberta['cohesion']],axis=1)
df1['average'] = df1.iloc[:, 0:4].mean(axis=1)
df1
```

```python
df2 = pd.DataFrame()
df2 = pd.concat([tfidf['syntax'],svr['syntax'],ngram['syntax'],roberta['syntax']],axis=1)
df2['average'] = df2.iloc[:, 0:4].mean(axis=1)
df2
```

```python
df3 = pd.DataFrame()
df3 = pd.concat([tfidf['vocabulary'],svr['vocabulary'],ngram['vocabulary'],roberta['vocabulary']],axis=1)
df3['average'] = df3.iloc[:, 0:4].mean(axis=1)
df3
```

```python
df4 = pd.DataFrame()
df4 = pd.concat([tfidf['phraseology'],svr['phraseology'],ngram['phraseology'],roberta['phraseology']],axis=1)
df4['average'] = df4.iloc[:, 0:4].mean(axis=1)
df4
```

```python
df5 = pd.DataFrame()
df5 = pd.concat([tfidf['grammar'],svr['grammar'],ngram['grammar'],roberta['grammar']],axis=1)
df5['average'] = df5.iloc[:, 0:4].mean(axis=1)
df5
```

```python
df6 = pd.DataFrame()
df6 = pd.concat([tfidf['conventions'],svr['conventions'],ngram['conventions'],roberta['conventions']],axis=1)
df6['average'] = df6.iloc[:, 0:4].mean(axis=1)
df6
```

```python
df_total = pd.DataFrame()
df_total = pd.concat([df1['average'],df2['average'],df3['average'],df4['average'],df5['average'],df6['average']],axis=1)
col = ['cohesion','syntax','vocabulary','phraseology','grammar','conventions']
df_total.columns = col
#df_total = df_total[0:3]
df_total
```

```python
y_real = pd.read_csv("data/validation.csv",index_col=0)
y_real = y_real[['cohesion','syntax','vocabulary','phraseology','grammar','conventions']]
```

```python
import math
row = 758
#row = 3
col = 6
total = 0
sum_total = 0
for i in range(col):
    for j in range(row):
        total += (y_real.iloc[j][i] - df_total.iloc[j][i])**2
    total /= row
    sum_total += math.sqrt(total)
    total = 0
sum_total/= col
sum_total
```
