```python
import pandas as pd
import re
```

```python
test = pd.read_csv("data/test.csv",index_col=0)
```

```python
test
```

```python
test['full_text'] = test["full_text"].replace(re.compile(r'[\n\r\t]'), ' ', regex=True)
```

```python
length = []
for i in test['full_text']:
    leng = len(i.split())
    length.append(leng)

test['text_length']=length

#30efb9040e0ec8b40219c7025888f89db4a60801
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
text_tag = []
for i in test['full_text']:
    i = word_tokenize(i)
    pos_tagger = nltk.pos_tag(i)
    word_tag = []
    for j in pos_tagger:
        (word,tag) = j
        word_tag.append(tag)
    text_tag.append(word_tag)
```

```python
import nltk
from nltk.data import load
nltk.download('tagsets')
tagdict = load('help/tagsets/upenn_tagset.pickle')
tagdict.keys()
```

```python
tags = list(tagdict.keys())
tags.append("#")
tags
```

```python
freq = pd.DataFrame(index=range(0,len(test)),columns=tags)
freq.iloc[:] = 0
freq = freq.astype('int')
```

```python
from collections import Counter
d = []
for i in text_tag:
    d.append(Counter(i))
```

```python
for i in range(len(test)):
    for key, value in d[i].items():
        freq[key][i] = value
```

```python
verb = ["VB","VBG","VBD","VBN","VBP","VBZ"]
wh = ["WDT","WP","WRB","PDT","DT","WP$"] #determiner
connection = ["UH","RP","TO","IN","CC","MD","EX"]
adb = ["RB","RBR","RBS"]
pronoun = ["POS","PRP","PRP$"]
noun = ["NN","NNS","NNP","NNPS","LS"]
adj = ["JJ","JJR","JJS"]
fw = ["FW"] #foreign word
number = ['CD'] #number
punc = ["''","--","(",")",".",",","``","$","#","SYM"]

word_type = [verb,wh,connection,adb,pronoun, noun,adj,fw,number,punc]
```

```python
test['verb'] = freq[verb].sum(axis=1)
test['wh'] = freq[wh].sum(axis=1)
test['connection'] = freq[connection].sum(axis=1)
test['adb'] = freq[adb].sum(axis=1)
test['pronoun'] = freq[pronoun].sum(axis=1)
test['noun'] = freq[noun].sum(axis=1)
test['adj'] = freq[adj].sum(axis=1)
test['fw'] = freq[fw].sum(axis=1)
test['number'] = freq[number].sum(axis=1)
test['punc'] = freq[punc].sum(axis=1)
```

```python
test = test[['text_id','full_text','text_length','verb','adj','adb','punc']]
```
